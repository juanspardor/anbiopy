# Functions:

## 1. taxonomy_df()

Creates a DataFrame for the taxonomic classification of each OTU, where each column represents one taxon. The main input is the taxonomic classifcation data of each OTU. Something important to consider is that it **must** contain the 'Taxonomy' column, all other columns are maintained and included in the output. 

**Parameters:**
- **taxonomic_data**: DataFrame that has a 'Taxonomy' column, which contains the taxonomic classification of the OTU using the following pattern: k__KingdomName;p__PhyluName;c__ClassName;o__OrderName;f__FamilyName;g__GenusName;s__SpeciesName

**Output:** A DataFrame with the complete OTU classification 

````python
taxo_df = ab.taxonomy_df(taxonomic_data = otu_taxonomy)
taxo_df.head()
````

## 2. relative_abundaces()

Calculates the relative abundances of each OTU in a OTU observation data frame. 

**Parameters:**
- **otus**: OTU observation DataFrame. It **must not** contain an ID column.

**Outpu**: A DataFrame with the relative abundance of each OTU for each sample in decimal form

````python
#Remove ID column from otu data
otu_data_no_id = otu_data.drop(columns = ['ID']).copy()

#Calculate relative abundances
rAbundances = ab.relative_abundances(otus = otu_data_no_id)
rAbundances.head()
````

## 3. rarefaction()

Performs rarefication on a dataset of OTU observations.

**Parameters:**
- **otus**: OTU observation DataFrame. It **must not** contain an ID column
- **seed (Optional)**: Random seed for replicability. Default value = 23

**Output**: The rarefied OTU dataset as a Numpy ndarray (a matrix)

````python
rarefied_otus = ab.rarefaction(otus = otu_data_no_id,
                              seed = 25)
rarefied_otus
````

## 4. alpha_diversity()

Calculates the alpha diversity of each individual for a given set of metrics.

**Parameters:**
- **otus**: OTU Observation DataFrame. It can include an ID column
- **metrics**: A String list containing the desired metrics. Supported metrics are: "shannon", "sv-richness", "heip_e", "simpson".
- **includes_id (Optional)**: Boolean that indicates if the pData includes an ID column. Default value = False
- **rarefy (Optional)**: Boolean that indicates if the user desires to perform rarefaction on the dataset before calculating alpha-diversities. Default value = True
- **seed (Optional)**: Random seed for replicability. Default value = 23

**Output**: A DataFrame containing each desired metric for each sample. If the user included an ID column it will also be included and named "Sample_ID"

````python
#Using the otu_data version that has an ID column. 
alpha_diversity_test = ab.alpha_diversity(otus = otu_data,
                                         metrics = ["shannon", "sv-richness", "heip_e", "simpson"],
                                         includes_ids = True,
                                         rarefy = True,
                                         seed = 24)

alpha_diversity_test.head()
````

## 5. beta_diversities()

Calculates the alpha diversity of each individual for a given set of metrics.

**Parameters:**
- **otus**: OTU Observation DataFrame. I can include an ID column
- **metrics**: A String list containing the desired metrics. Supported metrics are: *"weighted_unifrac"*, *"unweighted_unifrac"*.
- **tree_string**: String that represents the phylogenetic tree of the OTUs
- **includes_id (Optional)**: Boolean that indicates if the pData includes an ID column. Default value = False
- **rarefy (Optional)**: Boolean that indicates if the user desires to perform rarefaction on the dataset before calculating alpha-diversities. Default value = True
- **seed (Optional)**: Random seed for replicability. Default value = 23

**Output**: A list that in each position has the beta diversity values for its corresponding metric. Each position contains a list of the elements in the upper triangular of the NxN beta diveristy (these elements do not include the diagonal). If the complete matrix is needed, use *convert_vector_beta_matrix()*.

````python
beta_diversities_test = ab.beta_diversities(otus = otu_data, 
                                            tree_string = tree,
                                            metrics = ["weighted_unifrac", "unweighted_unifrac"],
                                            includes_ids = True,
                                            rarefy = True,
                                            seed = 24)
````

````python
#Weighted unifrac list
w_unifrac = beta_diversities_test[0]
w_unifrac
````

````python
#Unweighted unifrac list
unw_unifrac = beta_diversities_test[1]
unw_unifrac
````

## 6. convert_vector_beta_matrix()

Based on a list of elements, reconstructs the NxN symmetric matrix from which they come from. This matrix has 0's on the diagonal

**Parameters:**
- **vector**: List of elements

**Output**: A numpy (ndarray) NxN matrix with 0 in its diagonal. 

````python
#Reconstruct the weighted unifrac beta diversity matrix
beta_matrix = ab.convert_vector_beta_matrix(w_unifrac)
beta_matrix
````

## 7. convert_matrix_vector()

Given a NxN matrix, it creates a list of the elements in its upper triangular, not including the diagonal. This is the function that is used inside *beta_diversites()*

**Parametrs:**
- **matrix**: NxN matrix

**Output**: A numpy list with the elements of the matrix

````python
beta_vector = ab.convert_matrix_vector(beta_matrix)
beta_vector
````

## 8. create_graphic()

Creates different types of graphics depending on parameter input. The type of graphics are explained in the **graphic_type** parameter. 

**Parameters:**

- **X**: Observations of explicative variables
- **graphic_type**: Number indicating the type of graphic to be created.
  - 1: Y vs. Numeric X -> Line plot
  - 2: Y vs X pData -> Violin plot 
  - 3: Y vs Numeric X (first column) and Categoric X (second column) -> Multiple line plot
- **save_file (Optional)**: Boolean to indicate if the created graph should be saved. Default value = False
- **file_name (Optional)**: String for the file name in case the graphic should be saved. Default value = " "

**Output**: No return value. It prints the image on the cell and saves it if necessary

````python
#Test for graphic_type = 1
ab.create_graphic(Y = alpha_diversity_test['shannon'],
                 X = antro_data['age'],
                 graphic_type = 1,
                 save_file = True,
                 file_name = 'LinearTest')
````

## 9. anova_test()

Creates, runs and evaluates an ANOVA model using selected variables. It validates the 3 assumptions of the ANOVA model: Homocedasticity, Error Normality, Independence.
The function call works similar to *aov()* in R, in the sense that it can receive an explicative variable matrix and the user indicates which ones to use. 

**Parameters:**:
- **Y**: Observations of the objective variable. This should be a numeric variable
- **data**: DataFrame containing multiple variables.
- **X_columns**: List of strings containing the names of the variables in **data** to be used in the model. All these variables should be categoric.
- **alpha (Optional)**: Alpha value for statistical tests. Default value = 0.05

**Output**: No return value. It prints a message with the following information:
1. ANOVA model results
2. Homocedasticity test and result
3. Error Normality test and result
4. Independe test and result
5. Assumption conclusion.

    5.1 If all assumptions are validated -> Confirmation
    5.2 If at least one assumption is not validated -> Indication and possible transformation (depends on number of explicative variables)   

````python
# ONE-WAY ANOVA
ab.anova_test(Y = alpha_diversity_test['shannon'],
             data = antro_data,
             X_columns = ['sex'],
             alpha = 0.05)
````

````python
## Multiple explicative variables
ab.anova_test(Y = alpha_diversity_test['shannon'],
             data = antro_data,
             X_columns = ['sex', 'city'],
             alpha = 0.05)
````

## 10. multi_level_factors()

For a given objective variable, identifies possible important predictors at different taxonomic levels using Random Forest predicting models. It produces a grid of the most important predictors at each taxonomic level. 

Basic logic of execution:
1. For each selected taxonomic level:

    1.1 Groups all OTU observations based on their group within the taxonomic level
   
    1.2 Creates a RF model using the grouped OTU observations. Be default all RF models are created using sqrt(No. Columns) variables
   
    1.3 Selected the top **max_rows** groups and includes them in the level's plot.
3. Returns the list of importances for each level

**Parameters**:
- **otus**: DataFrame of OTU observations
- **otu_taxonomy**: DataFrame of taxonomic classification of each OTU. It must contain the same columns as the output of *taxonomy_df()* i.e., it must have Kingdom, Phylum,..., columns. 
- **Y**: Observations of the objective variable
- **Y_name**: Name of the objective variable, for the title of the plots.
- **levels**: List of strings containing the taxonomic levels to be used. Supported values are "Kingdom", "Phylum", "Class", "Order", "Family", "Genus", "Species"
- **classification (Optional)**: Boolean to indicate if the predicting task is classification or not. Default value = False
- **includes_ids (Optional)**: Boolean to indicate if **otus** contains an ID column. Default value = False
- **max_rows (Optional)**: Number of top predictors to include in each plot. Default value = 10
- **relatve (Optional)**: Boolean to indicate if relative abundances should be calculated for **otus**. Default value = True
- **seed (Optional)**: Randon seed for replicability of results. Default value = 23
- **save_file (Optional)**: Boolean to indicate if the final plot should be saved. Defalut value = False
- **file_name (Optional)**: String for the file name of the image should it be saved

**Ouputs**
- Prints the grid of plots. If indicated, saves the grid of plots. 
- Returns a list of DataFrames, each one corresponding to its respective pLevels value. Each DataFrame contains the complete table of Predictor-Importance values

````python
#Important factos for BMI (regression task) at levels Class, Order, Family, Genus
multiple_predictors_test = ab.multi_level_factors(otus = otu_data,
                                                    otu_taxonomy = taxo_df,
                                                    Y = antro_data['bmi'],
                                                    Y_name = "BMI", 
                                                    levels = ['Class', 'Order', 'Family', 'Genus'],
                                                    classification = False,
                                                    includes_ids = True, 
                                                    max_rows = 5, 
                                                    relative = True,
                                                    seed = 24, 
                                                    save_file = True,
                                                    file_name = "multiple_factor_test")
````

````python
#All predictors for Order Taxon (second inputed level)
multiple_predictors_test[1].head()
````

## 13. friedman_test()

Non-parametric statistical test. It tests if the distribution of an objective variable is the same between different groups. 

**Requirements:**
- Only 1 explicative variable
- All groups must have the same number of observations
- At least 3 groups

**Parameters:**
- **Y**: Observations of the objective variable. Must be numeric
- **X**: Observations of the explicative variable. Must be categoric
- **alpha (Optional)**: Alpha value for statistical tests. Defalut value = 0.05

**Outputs**: No return value. It prints the result of the test

````python
#Data preparation for same number of observations of each group
data_bog = antro_data[antro_data['city'] == 'Bogota'].head(50)
data_med = antro_data[antro_data['city'] == 'Medellin'].head(50)
data_baq = antro_data[antro_data['city'] == 'Barranquilla'].head(50)
friedman_data = pd.concat([data_bog, data_med, data_baq], axis = 0)

#Select the alpha diversity values for those individuals
friedman_alpha_data = alpha_diversity_test[alpha_diversity_test['Sample_ID'].isin(friedman_data['ID'])]
````

````python
#Friedmant test using data that complies with its requirements
ab.friedman_test(Y = friedman_alpha_data['shannon'],
                X = friedman_data['city'],
                alpha = 0.05)
````
